```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
grumps_plain_html = requests.get('https://www.youtube.com/user/GameGrumps/videos')
html_text = grumps_plain_html.text
```

```python
soup = BeautifulSoup(html_text)
print(soup.prettify())
```

```python
newest_vids = soup.find_all('h3',{"class": "yt-lockup-title"})[0:9]
```

Pretty simplistic in getting the objects containing the data that I am looking for (aka the name of the video, the href associated with it and a unique ID).
