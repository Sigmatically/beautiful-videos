## Achievement Hunter

Get an understanding of the achievement hunter website and the way new video content is displayed in order to add new videos

```python
response = requests.get('http://achievementhunter.roosterteeth.com/')
achievehtml = response.text
print(achievehtml)
```

```python
soup = BeautifulSoup(achievehtml, "html.parser")
print(soup.prettify())
```

Thought the Carousel HTML that displays recent videos would be as easy to navigate through as it was using the chrome web developer tools. However, it seems the javascript code behind the carousel takes in a comment with an array full of video objects storing video information and builds a moving carousel from it. This is probably done through react or some sort of other library. At first I thought it would be a hinderance but it actually makes it easier to iterate over.

```python
viddiv = soup.find(id="recent-carousel-comment")
print(str(viddiv.contents[1]))
```

Originally, I thought I may have been able to literally evaluate this into a list, but after many failed attempts I remembered Python has a great JSON parser exactly made for this. Because this is JSON data.

```python
import json
```

```python
video_list = json.loads(str(viddiv.contents[1]))
```
